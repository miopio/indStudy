```python
#clean the ForeName data: take out single character first names, first names plus single characters
import re

    
def getfirstname(word):    
    matchstr = r'\w{2,10000}'
    if isinstance(word, str) is False:
        return 'None'
    else:
        out = re.search(matchstr,word)
        if out is None:
            return 'None'
        return out.group()

names = [getfirstname(word) for word in authors['ForeName']]

# for word in authors['ForeName']:
#     if getfirstname(word) is None:
#         print(word)
    
```

```python
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```
